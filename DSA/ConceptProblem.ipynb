## Two-Pointer Pattern 
- When to use it:
- You are dealing with a sorted array or linked list.
- You need to find a set of elements (pairs, triplets) that satisfy a specific condition (e.g., sum to a target).
- You need to compare elements from opposite ends (e.g., checking for palindromes).
- Keywords: "Sorted array", "Pair that sums to...", "Reverse the array".

````python
# Template
def two_pointer_template(arr, target):
    left = 0
    right = len(arr)-1
    while left < right:
        current_sum = arr[left] + arr[right]
        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1
        else:
            right -=1
    return -1
````

### (GOOGLE) Problem LC 15: 3SUM (https://leetcode.com/problems/3sum/description/)  

````python
from typing import List
def threeSum(nums: List[int])-> List[List[int]]:
    res = []
    nums.sort()
    for i in range(len(nums)):
        if i>0 and nums[i]==nums[i-1]:
            continue
        left = i+1
        right = len(nums)-1
        while left<right:
            curr_sum = nums[i] + nums[left] + nums[right]
            if curr_sum == 0:
                res.append([nums[i], nums[left], nums[right]])
                left += 1
                right -= 1
                while left<right and nums[left] == nums[left-1]: #to not include the same sets (keeps the sets distinct)
                    left+=1
                while left<right and nums[right] == nums[right+1]:
                    right -= 1
            elif curr_sum > 0:
                right -= 1
            else:
                left += 1
    return res
````

## Stacks, Queues and Monotonic Patterns

````python
# MONOTONIC STACK TEMPLATE FOR INCREASING AND DECREASING
# In most LeetCode problems, you should store the index i in the stack rather than the value arr[i].

# Why? Because the index gives you two pieces of information for the price of one:

# The value: arr[stack[-1]]

# The distance: current_index - stack[-1] (Essential for "how many days" or "width" questions).

def monotonicIncreasing(arr):
    stack = []
    for i in range(len(arr)):
        # While stack is not empty and current element is SMALLER 
        # than the top, the 'increasing' rule is broken.
        while stack and arr[i] < stack[-1]:
            stack.pop()
        
        stack.append(arr[i])
    return stack

def monotonicDecreasing(arr):
    stack = []
    for i in range(len(arr)):
        # While stack is not empty and current element is LARGER 
        # than the top, the 'decreasing' rule is broken.
        while stack and arr[i] > stack[-1]:
            stack.pop()
        
        stack.append(arr[i])
    return stack
````

## Binary Search Mastery

#### 1. Exact match (the standard)

Use when you need **some** index where `nums[i] == target`. If duplicates exist, any matching index is fine.

**Key traits**

- Search range is inclusive on both ends: `[0, len(nums) - 1]`.
- Loop: `while l <= r`.
- Shrink so pointers can cross: `l = mid + 1`, `r = mid - 1`.

```python
def binary_search_exact(nums, target):
    l, r = 0, len(nums) - 1

    while l <= r:
        mid = l + (r - l) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            l = mid + 1
        else:
            r = mid - 1

    return -1  # target not found
```

#### 2. Lower bound (first `i` with `nums[i] >= target`)

Matches the usual “lower bound” / bisect-left idea: first position where the value is **not less than** `target`. Use for the **first occurrence** in a sorted array with duplicates, or for the **insertion index** that keeps order when `target` is missing.

**Key traits**

- `r` starts at `len(nums)` (not `len(nums) - 1`) so the answer can be “past the end” when every element is `< target`.
- Loop: `while l < r` (ends at a single index).
- When `nums[mid] >= target`, set `r = mid` (not `mid - 1`, because `mid` may be the answer).

```python
def binary_search_lower_bound(nums, target):
    l, r = 0, len(nums)  # note: len(nums), not len(nums) - 1

    while l < r:
        mid = l + (r - l) // 2

        if nums[mid] >= target:
            r = mid
        else:
            l = mid + 1

    return l
```

#### 3. Upper bound (first `i` with `nums[i] > target`)

Same half-open setup as lower bound. Use for the **first element strictly greater** than `target`, or derive **last occurrence** of `target` as `upper_bound_index - 1` (with bounds checks).

**Key traits**

- Same as lower bound: `l = 0`, `r = len(nums)`, `while l < r`.
- Only change: branch on `nums[mid] > target` instead of `>=`.

```python
def binary_search_upper_bound(nums, target):
    l, r = 0, len(nums)

    while l < r:
        mid = l + (r - l) // 2

        if nums[mid] > target:
            r = mid
        else:
            l = mid + 1

    return l
```

### Problem LC 236: LCA of BT (https://leetcode.com/problems/lowest-common-ancestor-of-a-binary-tree/description/)

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if not root or root==p or root == q:
            return root

        left_res = self.lowestCommonAncestor(root.left, p, q)
        right_res = self.lowestCommonAncestor(root.right, p, q)
        if left_res and right_res:
            # found lca
            return root
        return left_res if left_res else right_res
````

## GRAPH (BFS/DFS)

#### BFS DFS Template for Tree and Graph

````python
# TREE Template
class TreeNode:
    def __init__(self, val = 0, left= None, right = None):
        self.val = val
        self.left = left
        self.right = right
    
    def dfs_tree(root):
        # Base case: if the node is null, stop and return
        if not root:
            return
        
        # 1. PRE-ORDER: Process node before children
        # print(root.val) 
        
        dfs_tree(root.left)
        
        # 2. IN-ORDER: Process node between left and right children (sorts BSTs)
        # print(root.val)
        
        dfs_tree(root.right)
        
        # 3. POST-ORDER: Process node after children (useful for bottom-up tasks)
        # print(root.val)

    def bfs(root):
        if not root:
            return
    
        # Initialize queue with the root node
        queue = deque([root])
        level = 0
        
        while queue:
            # Get the number of nodes at the current level
            level_size = len(queue) 
            
            # Process all nodes at the current level
            for _ in range(level_size):
                node = queue.popleft()
                
                # Process the current node
                # print(node.val)
                
                # Add children to the queue for the next level
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                    
            # Increment level after processing all nodes on it
            level += 1
````

````python
# GRAPH Template
def dfs_graph(graph, start_node, visited=None):
    # Initialize the visited set on the first call
    if visited is None:
        visited = set()
        
    # Mark the current node as visited
    visited.add(start_node)
    
    # Process the node
    # print(start_node)
    
    # Explore unvisited neighbors
    for neighbor in graph[start_node]:
        if neighbor not in visited:
            dfs_graph(graph, neighbor, visited)
            
    return visited


from collections import deque

def bfs_graph(graph, start_node):
    # Initialize queue and visited set
    queue = deque([start_node])
    visited = set([start_node]) 
    
    while queue:
        # Pop the front node
        node = queue.popleft()
        
        # Process the node
        # print(node)
        
        # Check all neighbors
        for neighbor in graph[node]:
            if neighbor not in visited:
                # Mark as visited BEFORE adding to queue
                visited.add(neighbor)
                queue.append(neighbor)
````

````python
def findRedundantConnection(edges):
    n = len(edges)
    rank = [1]*(n+1)
    parent = [i for i in range(n+1)]

    def find(n):
        if parent[n]!=n:
            parent[n] = find(parent[n])
        return parent[n]
    
    def union(n1, n2):
        p1, p2 = find(n1), find(n2)
        if p1==p2:
            return False
        
        if rank[p1] > rank[p2]:
            parent[p2] = p1
            rank[p1]+= rank[p2]
        else:
            parent[p1] = p2
            rank[p2]+= rank[p1]
        
        return True

    for u, v in edges:
        if not union(u, v):
            return [u,v]
````

### Min_heap from scratch (shift_up and shift_down)

````python
class MinHeap:
    def __init__(self):
        self.heap = []

    def insert(self, val: int):
        """Adds a new value to the heap."""
        # 1. Add the new element to the very end of the array
        self.heap.append(val)
        # 2. Sift it up into its correct position
        self._sift_up(len(self.heap) - 1)

    def extract_min(self) -> int:
        """Removes and returns the smallest value (the root)."""
        if not self.heap:
            return None
        if len(self.heap) == 1:
            return self.heap.pop()

        # 1. Save the minimum value (the root) to return later
        min_val = self.heap[0]
        # 2. Move the very last element in the array to the root position
        self.heap[0] = self.heap.pop()
        # 3. Sift this new root down into its correct position
        self._sift_down(0)

        return min_val

    def _sift_up(self, index: int):
        """
        Moves a node up the tree as long as it is smaller than its parent.
        Used after inserting a new element at the bottom.
        """
        parent_index = (index - 1) // 2

        # While we haven't reached the root, AND the current node is smaller than its parent
        while index > 0 and self.heap[index] < self.heap[parent_index]:
            # Swap the current node with its parent
            self.heap[index], self.heap[parent_index] = self.heap[parent_index], self.heap[index]
            
            # Update indices to continue moving up the tree
            index = parent_index
            parent_index = (index - 1) // 2

    def _sift_down(self, index: int):
        """
        Moves a node down the tree by swapping it with its smallest child.
        Used after extracting the root and replacing it with a leaf.
        """
        length = len(self.heap)

        while True:
            left_child_idx = 2 * index + 1
            right_child_idx = 2 * index + 2
            smallest = index # Assume the current node is the smallest for now

            # If the left child exists and is smaller than the current node
            if left_child_idx < length and self.heap[left_child_idx] < self.heap[smallest]:
                smallest = left_child_idx

            # If the right child exists and is smaller than the currently known smallest
            if right_child_idx < length and self.heap[right_child_idx] < self.heap[smallest]:
                smallest = right_child_idx

            # If the smallest node is still the current node, the heap property is satisfied
            if smallest == index:
                break

            # Otherwise, swap the current node with the smallest child
            self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
            
            # Update the index to continue moving down the tree
            index = smallest
````

### Problem LC 1046: Last Stone Weight (https://leetcode.com/problems/last-stone-weight/description/)

````python
def lastStoneWeight(self, stones: List[int]):
    hq = [-x for x in stones]
    heapq.heapify(hq)
    while len(hq)>1:
        first = -heapq.heappop(hq)
        second = -heapq.heappop(hq)
        heapq.heappush(hq, -(first - second))
    return -hq[0] if hq else 0
````

### GOOGLE Problem LC 973: K Closest Points to Origin (https://leetcode.com/problems/k-closest-points-to-origin/description/)

````python
class Solution:
    def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
        max_heap = []
        heapq.heapify(max_heap)
        for point1, point2 in points:
            dist = point1**2 + point2**2
            heapq.heappush(max_heap, (-dist, point1, point2))
            if len(max_heap)>k:
                heapq.heappop(max_heap)
        
        return [[x,y] for i, x, y in max_heap]
````

````python
class MedianFinder:
    def __init__(self):
        self.upper_half = [] # min_heap
        self.lower_half = [] #max_heap

    def addNum(self, num):
        # 1. Push to lower half (max-heap)
        heapq.heappush(self.lower_half, -num)

        # 2. Ensure every element in lower_half <= every element in upper_half
        if (self.lower_half and self.upper_half and 
            (-self.lower_half[0] > self.upper_half[0])):
            val = -heapq.heappop(self.lower_half)
            heapq.heappush(self.upper_half, val)

        # 3. Handle size imbalance (if diff > 1)
        if len(self.lower_half) > len(self.upper_half) + 1:
            val = -heapq.heappop(self.lower_half)
            heapq.heappush(self.upper_half, val)
        elif len(self.upper_half) > len(self.lower_half) + 1:
            val = heapq.heappop(self.upper_half)
            heapq.heappush(self.lower_half, -val)

    def findMedian(self):
        if len(self.lowerhalf) > len(self.upperhalf):
            return -self.lowerhalf[0]
        elif len(self.upperhalf)>len(self.lowerhalf):
            return self.upperhalf[0]
        else:
            return (-self.lowerhalf[0]+self.upperhalf[0])/2
````

### GOOGLE : PROBLEM LC 212: Word Search 2 (https://leetcode.com/problems/word-search-ii/)

````python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        root = TrieNode()
        for word in words:
            node = root
            for ch in word:
                if ch not in node.children:
                    node.children[ch] = TrieNode()
                node = node.children[ch]
            node.word = word
        
        m, n = len(board), len(board[0])
        res = []

        def dfs(r, c, node):
            if r<0 or r>=m or c<0 or c>=n or board[r][c] not in node.children:
                return
            char = board[r][c]
            child = node.children[char]

            if child.word is not None:
                res.append(child.word)
                child.word = None
            
            board[r][c] = "#"
            dfs(r+1, c, child)
            dfs(r-1, c, child)
            dfs(r, c+1, child)
            dfs(r, c-1, child)

            board[r][c] = char

            if not child.children:
                del node.children[char]


        for i in range(m):
            for j in range(n):
                dfs(i, j, root)
            
        return res


````

### Problem LC 198: House Robber  (https://leetcode.com/problems/house-robber/description/)

````python
class Solution:
    def rob(self, nums: List[int]) -> int:
        # n = len(nums)
        # mem= [-1]*(n+1)
        # def recurse(n):
        #     if n<=0:
        #         return 0
        #     if n==1:
        #         return nums[0]
        #     if mem[n]!=-1:
        #         return mem[n]
        #     mem[n] = max(recurse(n-1), recurse(n-2) + nums[n-1])
        #     print(mem)
        #     return mem[n]
        
        # return recurse(n)
        n = len(nums)
        dp = [0]*(n+1)
        dp[0] = 0
        dp[1] = nums[0]
        for i in range(2, n+1):
            dp[i] = max(dp[i-1], nums[i-1]+ dp[i-2])
        return dp[n]
````

````python
class Solution:
    def rob(self, nums: list[int]) -> int:
        # Initialize our two variables
        rob1 = 0
        rob2 = 0
        
        # [rob1, rob2, n, n+1, ...]
        for n in nums:
            # Evaluate the current house
            # Choice 1: Rob current (n) + max from two houses ago (rob1)
            # Choice 2: Skip current, keep max from one house ago (rob2)
            current_max = max(rob1 + n, rob2)
            
            # Slide the window forward for the next iteration
            rob1 = rob2
            rob2 = current_max
            
        # By the end of the loop, rob2 holds our final maximum
        return rob2
````

### Problem LC 300: Longest Increasing sub-sequence (https://leetcode.com/problems/longest-increasing-subsequence/description/)

````python
class Solution: # DP soln o(n^2)
    def lengthOfLIS(self, nums):
        dp = [1] * len(nums)
        for i in range(len(nums)):
            for j in range(i):
                if nums[j]<nums[i]:
                    dp[i] = max(dp[i], 1+dp[j])
        return max(dp)
````

````python
# Binary search solution O(nlogn)
from bisect import bisect_left

class Solution:
    def lengthOfLIS(self, nums: list[int]) -> int:
        # This list will store our active subsequence
        sub = []
        
        for num in nums:
            # If the list is empty OR num is greater than the last element
            if len(sub) == 0 or num > sub[-1]:
                sub.append(num)
            else:
                # Binary search to find the index of the first element >= num
                # bisect_left does exactly this!
                replace_index = bisect_left(sub, num)
                
                # Replace the element to create a better "anchor" for the future
                sub[replace_index] = num
                
        # The length of our 'sub' list is the length of the LIS
        # (Note: the actual elements in 'sub' might NOT be the valid sequence itself)
        return len(sub)
````
